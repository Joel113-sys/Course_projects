```python
import pandas as pd
import os
import warnings

warnings.filterwarnings("ignore")
```

```python
def csv_to_pkl(file_path, drop_rows=3):
    df = pd.read_csv(file_path, header=None, skiprows = drop_rows)
    df.drop(columns=[df.columns[0]], inplace=True)
    if "Y" in file_path:
        if "task1" in file_path:
            pickle_path = "./data/" + file_path.split("/")[-1].replace(".csv", "_1.pkl")
        else:
            pickle_path = "./data/" + file_path.split("/")[-1].replace(".csv", "_2.pkl")
    else:
        pickle_path = "./data/" + file_path.split("/")[-1].replace(".csv", ".pkl")
    df.to_pickle(pickle_path)
    return
```

```python
if not os.path.isdir('./data'):
    os.mkdir('./data') 

csv_to_pkl("./stat3612-project-task1/X_train.csv", 3)
csv_to_pkl("./stat3612-project-task1/X_valid.csv", 3)
csv_to_pkl("./stat3612-project-task1/X_test.csv", 3)
csv_to_pkl("./stat3612-project-task1/Y_train.csv", 1)
csv_to_pkl("./stat3612-project-task1/Y_valid.csv", 1)
csv_to_pkl("./stat3612-project-task2/Y_train.csv", 1)
csv_to_pkl("./stat3612-project-task2/Y_valid.csv", 1)
```
